```python
import pandas as pd
```

```python
user_df = pd.read_csv("./user_data.csv")
logs_df = pd.read_csv("./logs.csv")

logs_shape = logs_df.shape
user_shape = user_df.shape
print(f"User shape df: {logs_shape}, Logs shape df: {user_shape}")
```

```python
df = logs_df.merge(user_df, on="client", how="inner")
df
```

```python
# Какой клиент совершил больше всего успешных операций? (success == True)
successed_operations_clients = logs_df.query("success == True"). \
                         groupby("client", as_index=False). \
                         agg({"platform": "count"}). \
                         rename(columns={"platform" : "success_number"}). \
                         sort_values("success_number", ascending=False)

max_success_number = successed_operations_clients.success_number.max()
usefulest_clients = successed_operations_clients.query("success_number == @max_success_number").client.tolist()
usefulest_clients
```

```python
# С какой платформы осуществляется наибольшее количество успешных операций?

most_popular_platform = logs_df.query("success == True"). \
                                groupby("platform", as_index=False). \
                                agg({"client": "count"}). \
                                sort_values("client", ascending=False). \
                                iloc[0]

most_popular_platform
```

```python
# Какую платформу предпочитают премиумные клиенты?

most_popular_platform_among_premium_clients = df.query("premium == True"). \
                                                groupby("platform", as_index=False). \
                                                agg({"client": "count"}). \
                                                sort_values("client", ascending=False). \
                                                iloc[0]

most_popular_platform_among_premium_clients
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Визуализируйте распределение возраста клиентов в зависимости от типа клиента (премиум или нет)

fig, ax = plt.subplots(nrows=2, ncols=1)

sns.distplot(df.query("premium == False").age, ax=ax[0])
sns.distplot(df.query("premium == True").age, ax=ax[1])
```

```python
# Постройте график распределения числа успешных операций
clients_vs_successed_operations = successed_operations_clients.drop("client", axis=1)
sns.distplot(clients_vs_successed_operations, kde=False)
```

```python
# Визуализируйте число успешных операций, сделанных на платформе computer, в зависимости от возраста,
# клиенты какого возраста совершили наибольшее количество успешных действий?

success_operations_clients_by_compyter = df.query("platform == 'computer'"). \
                                            groupby("age", as_index=False). \
                                            agg({"success" : "sum"}). \
                                            sort_values("success", ascending=False)

plt.figure(figsize=[14, 8])
sns.barplot(x=success_operations_clients_by_compyter.age, y=success_operations_clients_by_compyter.success, color="gray")

print(f"Клиенты с возрастом {success_operations_clients_by_compyter.iloc[0].age} лет совершили наибольшее кол-во успешных действий")
```
